```python
#SEPARATE X and Y (X-independent attribute, Y-dependent)
Y=data['target']
```

```python
Y
```

```python
#X
X=data.drop(['target'],axis=1)
```

```python
X
```

```python
#DATA SPLITTING
#TRAIN-80%
#TEST-20%
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
#MODEL SELECTION- RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
```

```python
#create the model
model=RandomForestClassifier()
```

```python
#Train the model
model.fit(X_train,Y_train)
```

```python
#Model predictions
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
Y_pred.shape
```

```python
#Model evaluation
from sklearn.metrics import accuracy_score
```

```python
print('Accuracy on heart disease is',accuracy_score(Y_pred,Y_test)*100)
```
